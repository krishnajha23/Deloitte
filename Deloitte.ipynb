```python
import pandas as pd
import matplotlib.pyplot as plt
npscosts = pd.read_excel("NPSCOSTS.xlsx",sheet_name="Shortlisted Candidates")[1:]
potentialdf = pd.read_excel("Deloitte x UVA Systems Engineering Case Competition Data.xlsx", sheet_name="5 Parks for Selection Data").fillna(-1)
currentdf = pd.read_excel("Deloitte x UVA Systems Engineering Case Competition Data.xlsx", sheet_name="63 National Parks Data").fillna(-1)
currentdf = currentdf[currentdf['State'] != 'AK']
npscosts = npscosts.rename(columns={"Operation of the National Park System Organizations": "Park"})
finaldf = pd.merge(npscosts,currentdf,on='Park')
potentialfinaldf = pd.merge(npscosts,potentialdf,on='Park')
potentialfinaldf.head()
```

```python
import plotly.express as px
# color="Park", title = "Economic Output Based on Size"
potentialfinaldf['ratio'] = potentialfinaldf['FY 2023\nEnacted']/potentialfinaldf['Size (Gross Area Acres) ']
potentialfinaldf['ratio'] = potentialfinaldf['ratio'].round(2)
fig = px.scatter(finaldf, x="Size (Gross Area Acres) ",y="FY 2023\nEnacted",trendline = "ols")
fig.add_traces(
    px.scatter(potentialfinaldf, x="Size (Gross Area Acres) ", y="FY 2023\nEnacted",text='ratio').update_traces(marker_size=20, marker_color="red").data
)
fig.update_traces(marker_size=10)
#fig.update_layout(scattermode="group")
fig.update_layout(
yaxis_title="Cost (in thousands of dollars)"
)
fig.update_layout(
    font_family="Open Sans",
    #font_weight="bold",
    title="Operating Cost Based on the Size of the Park"
).show()
```

```python
fig = px.scatter(finaldf, x="FY 2023\nEnacted", y="Economic Output ($)", trendline='ols', title='Operation Cost vs. Economic Output')
fig.add_traces(
    px.scatter(potentialfinaldf, x="FY 2023\nEnacted", y="Economic Output ($)").update_traces(marker_size=20, marker_color="red").data
)
fig.update_layout(
xaxis_title="Cost (in thousands of dollars)"
)
fig.update_traces(marker_size=10)
fig.show()
```
